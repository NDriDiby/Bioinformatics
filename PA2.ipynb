```python
import re
#Get the number of motif to check
def motif_size(file):
'''this function will return 
    the size of the file'''
    f = open(file,'r')
    f = f.readlines()
    size = len(f)
    return size


#Open the file and read
def motif(file):
    '''This function will read a file
    line by line,you will need to pass the 
    path of the file as an argument'''
    f = open(file,'r')
    f = f.readlines()
    return f


#Clean the motif by removing line escape by empty space
def clean_motif(file):
    '''This function will return a clean
    sequence file by removing the empty space'''
    f = open(file,'r')
    f = f.read()
    dna= re.sub(r'\n','',f)
    return dna.upper()


#find motif for a file with a multiple lines
def find_motif(file1,file2):
    ''' 
    This function will find the motif from two file,
    first pass the file with the full motif
    then after pass the partten file
    you whish to find in file one.
    '''
    mot2_size = motif_size(file2)
    mot2 = motif(file2)
    mot1 = clean_motif(file1)
    for i in range(0,mot2_size):
        if('\n' in mot2[i]):
            clean = re.sub(r'\n','',mot2[i])
        else:
            clean = mot2[i]
        
        motifs = re.finditer(clean,mot1)
        
        for moti in motifs:
            gr = moti.group()
            position = moti.start()
            print(f"Motif {gr} found at position {position +1}")
            


def main():

    file1 = input("Copy and paste the path of your sequence file")
    file2 = input("Copy and paste the path of your parttern file")
    find_motif(file1,file2)


    file1 = input("Copy and paste the path of your sequence file")
    file2 = input("Copy and paste the path of your parttern file")
    find_motif(file1,file2)
    
  
    file1 = input("Copy and paste the path of your sequence file")
    file2 = input("Copy and paste the path of your parttern file")
    find_motif(file1,file2)
    
 
    file1 = input("Copy and paste the path of your sequence file")
    file2 = input("Copy and paste the path of your parttern file")
    find_motif(file1,file2)
    
if __name__ == '__main__':
    main()
```
